```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./frozen.csv")

plt.figure(figsize=(10, 6))
plt.plot(df["global_steps"], df["train/loss"], label="Train Loss", marker='o')
plt.plot(df["global_steps"], df["eval/loss"], label="Validation Loss", marker='x')

plt.xlabel("Global Steps")
plt.ylabel("Loss")
plt.title("Frozen Vision Tower: Train vs Validation Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
