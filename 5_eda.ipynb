# Exploratory Data Analysis

This notebook is used to:
- See distribution of each labels: no fire, dangerous fire, controlled fire
- to be decided ...

```python
import pandas as pd
df = pd.read_csv("fire_captions_combined.csv")
df
```

```python
label_counts = df['label'].value_counts()
print("Label Distribution:")
print(label_counts)
```

```python
import matplotlib.pyplot as plt

label_counts = df['label'].value_counts()
labels = label_counts.index
counts = label_counts.values
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# --- Bar chart ---
bars = ax1.bar(labels, counts, color=['green', 'red', 'orange'])
ax1.set_title('Label Distribution (Bar Chart)')
ax1.set_xlabel('Labels')
ax1.set_ylabel('Number of Images')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Add exact counts on top of each bar
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height + 1, f'{int(height)}',
             ha='center', va='bottom', fontsize=10)

# --- Pie chart ---
ax2.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90,
        colors=['lightgreen', 'tomato', 'gold'],
        wedgeprops={'edgecolor': 'black'})
ax2.set_title('Label Proportion (Pie Chart)')

# Tidy layout
plt.tight_layout()
plt.show()
```
