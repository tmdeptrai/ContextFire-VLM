```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

# Directory containing the annotated images
ANNOTATED_DIR = "fire_annotated"

# List all image paths
image_files = [os.path.join(ANNOTATED_DIR, f)
               for f in os.listdir(ANNOTATED_DIR)
               if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

def show_random_batch(batch_size=4):
    batch = random.sample(image_files, batch_size)
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))

    for i, ax in enumerate(axs.flat):
        img = Image.open(batch[i])
        ax.imshow(img)
        ax.set_title(os.path.basename(batch[i]), fontsize=10)
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Show a random batch of 4 images (2x2)
show_random_batch()
```
