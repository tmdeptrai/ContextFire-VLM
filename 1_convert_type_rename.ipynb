# Convert Types and Rename
This notebook is used to:
- Convert all images types (.png, .jpg, ~~.avif~~, .jpeg, .webp) to .jpg for a unified dataset
- Rename them by numerical index like image1.jpg, image2.jpg, ...

**CAUTION**: There is trouble with converting .avif images but I'm too lazy to fix it so let's just not have .avif images in your image folder

```python
IMG_FOLDER = "new_images"
```

```python
from PIL import Image
import os
from pathlib import Path

output_dir = Path(IMG_FOLDER)
output_dir.mkdir(exist_ok=True)

# Get list of supported image files
valid_extensions = ('.png', '.jpg', '.jpeg', '.webp', '.avif')
image_files = [f for f in os.listdir(IMG_FOLDER) 
               if f.lower().endswith(valid_extensions)]

starting_index = 1216

# Process each image
for idx, img_file in enumerate(image_files, start=starting_index):
    try:
        # Open and convert image
        img_path = os.path.join(IMG_FOLDER, img_file)
        with Image.open(img_path) as img:
            # Convert to RGB if necessary (for PNG with transparency)
            if img.mode in ('RGBA', 'P'):
                img = img.convert('RGB')
            
            # Save as JPG with new name
            new_name = f'image{idx}.jpg'
            new_path = os.path.join(IMG_FOLDER, new_name)
            img.save(new_path, 'JPEG')
            
            # Remove original if it's different from the new file
            if img_file != new_name:
                os.remove(img_path)
                
    except Exception as e:
        print(f"Error processing {img_file}: {str(e)}")

print(f"Converted {len(image_files)} images to JPG format")
```
