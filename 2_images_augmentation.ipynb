# Images Augmentation
This notebook is used to:
- Augment the image folder by introducing various types of image transformations, such as: Hue, Saturation, Brightness, Exposure, Blur,..
- Sample each images by x times

```python
IMG_FOLDER = "fire_dataset_3"
num_samples = 5
```

```python
import os
import cv2
import numpy as np

from albumentations import (
    Compose, HueSaturationValue, RandomBrightnessContrast,
    GaussianBlur, MotionBlur, HorizontalFlip,
)

transform = Compose([
    # Lighting and color variation
    HueSaturationValue(hue_shift_limit=10, sat_shift_limit=15, val_shift_limit=10, p=0.7),
    RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.7),

    # Blur to simulate camera softness
    GaussianBlur(blur_limit=(3, 5), p=0.3),
    MotionBlur(blur_limit=3, p=0.3),

    # Minor geometric flip
    HorizontalFlip(p=0.5),
])
```

```python
def augment_images(img_folder, num_samples=5):
    output_dir = img_folder + '_augmented'
    os.makedirs(output_dir, exist_ok=True)
    
    # Process each image in the folder
    for filename in os.listdir(img_folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Read image
            image_path = os.path.join(img_folder, filename)
            image = cv2.imread(image_path)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            
            # Generate augmented samples
            for i in range(num_samples):
                # Apply augmentation
                augmented = transform(image=image)
                aug_image = augmented['image']
                print(f"Successfully applied augmentation on {image_path}")
                
                # Save augmented image
                base_name = os.path.splitext(filename)[0]
                aug_filename = f"{base_name}_aug_{i+1}.jpg"
                aug_path = os.path.join(output_dir, aug_filename)
                
                # Convert back to BGR for saving
                aug_image = cv2.cvtColor(aug_image, cv2.COLOR_RGB2BGR)
                cv2.imwrite(aug_path, aug_image)
    
    print(f"Augmentation complete. Augmented images saved in {output_dir}")

# Apply augmentation to the image folder
augment_images(IMG_FOLDER, num_samples=num_samples)
```
